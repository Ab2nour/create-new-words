```python
from collections import defaultdict
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
def create_graph_from_words(words: list[str]) -> defaultdict:
    graph = defaultdict(lambda: defaultdict(int))

    for word in words:
        for i in range(len(word) - 1):
            letter_current = word[i]
            letter_after = word[i+1]
            graph[letter_current][letter_after] += 1

    return graph
```

```python
g = nx.DiGraph(graph)
```

```python
def draw_weighted_graph(graph):
    G = nx.DiGraph()  # Graphe orienté

    for node, edges in graph.items():
        for neighbor, weight in edges.items():
            G.add_edge(node, neighbor, weight=weight)

    pos = nx.spring_layout(G)  # Positionnement des nœuds
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=50, font_size=10, edge_color='gray')

    labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=10, label_pos=0.7)

    plt.show()
```

```python
draw_weighted_graph(graph)
```

```python
def choose_next_letter(letters_dict: dict[str, int]) -> str:
    next_letters = list(letters_dict.keys())
    frequences = list(letters_dict.values())

    probabilities = frequences / np.sum(frequences)
    next_letter = np.random.choice(next_letters, p=probabilities)

    return next_letter
```

```python
def create_word(letters_graph: defaultdict, start_letter: str = "a", length: int = 5) -> str:
    word = f"{start_letter}"
    current_letter = start_letter

    for i in range(length - 1):
        current_letter = choose_next_letter(letters_graph[current_letter])
        word += current_letter

    return word
```
